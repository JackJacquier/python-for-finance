# Plotting multiple functions in the same graph:
#### `plt` behaves like an object that holds information about anything we want to plot. If we want to plot multiple curves in the same image we can call `plt.plot()` multiple times. 

#### **Note:** whenever you call `plt.show()` the figure will display in a cell and plt. will reset

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,20,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.plot(x,z)
plt.show() #This will display the first plot with 2 curves
plt.plot(x,z)
plt.show() # this will display only one curve
```

# Figure objects
Figure objects are used to manipulate figure metadata, like the size, or background color (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html ) .The function `plt.figure()` returns a figure objects

* A Figure is the top-level container for all plot elements.

* Think of it as the canvas on which one or more plots (called Axes) are drawn.

```python
fig = plt.figure(figsize=(12,6), facecolor='lightskyblue')
fig.suptitle('Figure')
plt.plot(x,y)
plt.show()
```

#### Usually the main thing to care about is the size of the plot that can be set using `figsize=(width, height)` in inches

# Axis Objects

### Axis ojects hold the information about the actual plot.


#### **TIP:** Using an analogy with a painting and a frame; one can think of Figure objects as a frame that hold a painting and one can think of axis objects as the actual painting

### We can add an axis to an existing figure using `axis =figure.add_subplot()` method. Then we can use `axis.plot()` method

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot()
ax.plot(x,y)
ax.plot(x,z)
plt.show()
```

#### If you are doing a single plot, most of the times you won't care about the axis and can use just `plt`. Axis becomes useful when doing multiple plots in the same figure

# Useful plot settings to learn

#### * `plt.title("my_title",fontsize=18,pad=1.05)`: `fontsize` is the size of the title and `pad` is the distance of the title from the top of the chart

#### * `plt.plot(x,y,label="my_label")`: `label` is the name we want to use for each plot

#### * `plt.legend(fontsize=13)`: `fontsize` is the size of the labels we want to show

#### * `plt.grid(True)`: Shows a grid in the grahp

#### * `plt.xlabel("x_label",fontsize=15,labelpad=5)`: Add a title to the x axis with fontsize and distance from the axis given by `labelpad`

#### * `plt.ylabel("x_label",fontsize=15,labelpad=5)`: Add a title to the y axis with fontsize and distance from the axis given by `labelpad`

## **Advise:** Use large enough fontsize so that text is legible!

### How to change the default parameters

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,20,100)
y=np.sin(x)
z=np.cos(x)

fig = plt.figure(figsize=(12,6))
plt.plot(x,y,label="sin")
plt.plot(x,z,label="cos")
plt.title("my title",fontsize=18,pad=10)
plt.legend(fontsize=13)
plt.grid(True)
plt.xlabel("x_label",fontsize=15,labelpad=5)
plt.ylabel("y_label",fontsize=15,labelpad=5)
plt.show()
```

### **Important Remark:** Any plot you report in your coursework/thesis should at least contain the additional parameters above with fontsizes large enough that the picture becomes easy to read

## Xticks and Yticks: This settings allow us to control the size and the number of X and Y axis ticks we display

### * `plt.xticks(array_of_x_ticks,array_of_labels,fontsize=10)` here array_of_labels is optional in case we want to display different labels to numbers

### * `plt.ticks(array_of_y_ticks,array_of_labels,fontsize=10)`

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,20,100)
y=np.sin(x)
z=np.cos(x)

fig = plt.figure(figsize=(12,6))
plt.plot(x,y,label="sin")
plt.plot(x,z,label="cos")
plt.title("my title",fontsize=18,pad=10)
plt.legend(fontsize=13)
plt.grid(True)
plt.xlabel("x_label",fontsize=15,labelpad=5)
plt.ylabel("y_label",fontsize=15,labelpad=5)
plt.xticks([0,5,10,15,20],fontsize=10)
plt.yticks([-1,0,1],fontsize=10)
plt.show()
```

### We can also add labels (see below)

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,20,100)
y=np.sin(x)
z=np.cos(x)

fig = plt.figure(figsize=(12,6))
plt.plot(x,y,label="sin")
plt.plot(x,z,label="cos")
plt.title("my title",fontsize=18,pad=10)
plt.legend(fontsize=13)
plt.grid(True)
plt.xlabel("x_label",fontsize=15,labelpad=5)
plt.ylabel("y_label",fontsize=15,labelpad=5)
plt.xticks([0,5,10,15,20],["Jan","Feb","Mar","Apr","May"],fontsize=10)
plt.yticks([-1,0,1],[-10,0,10],fontsize=10)
plt.show()
```

# Saving figures to files

### `plt.savefig("filename',dpi=300)`, (dpi is the resolution of the pictue) typically this needs to be preceeded by `plt.tight_layout()` to make sure the output image gets the right figure/axis

### One can specify the format of the desired image file jpeg/png/eps in the filename extension


```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,20,100)
y=np.sin(x)
z=np.cos(x)

fig = plt.figure(figsize=(12,6))
plt.plot(x,y,label="sin")
plt.plot(x,z,label="cos")
plt.title("my title",fontsize=18,pad=10)
plt.legend(fontsize=13)
plt.grid(True)
plt.xlabel("x_label",fontsize=15,labelpad=5)
plt.ylabel("y_label",fontsize=15,labelpad=5)
plt.xticks([0,5,10,15,20],["Jan","Feb","Mar","Apr","May"],fontsize=10)
plt.yticks([-1,0,1],[-10,0,10],fontsize=10)
plt.tight_layout()
plt.savefig("filename.jpeg",dpi=300)
plt.show()
```

# Plotting time-series

```python
import yfinance
import pandas as pd
AAPL = yfinance.download(["AAPL"],period="5Y")
AAPL=pd.DataFrame(AAPL.values,AAPL.index,columns=['Close','High','Low','Open', 'Volume'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(12,6))
plt.plot(AAPL.index,AAPL.Close,label="AAPL")
plt.title("AAPL stock price",fontsize=18,pad=10)
plt.legend(fontsize=13)
plt.grid(True)
plt.xlabel("Date",fontsize=15,labelpad=5)
plt.ylabel("Price",fontsize=15,labelpad=5)
#plt.xticks([0,5,10,15,20],["Jan","Feb","Mar","Apr","May"],fontsize=10)
#plt.yticks([-1,0,1],[-10,0,10],fontsize=10)
plt.show()
```

#### One of the main things you may want to do is format the Xtick or Ytick values based on the values. For example instead of 200 we may want to display $200, so that is clear that the currency is USD

#### To do so we can use `tick.FuncFormatter()` from the submodule `import matplotlib.ticker as tick`

#### `tick.FuncFormatter(func)` takes a function as argument that can be used to format any array. This in turn can be applied to an axis object using a `axis.yaxis.set_major_formatter(tick.FuncFormatter(func))`

#### As we discussed, in order to use an axis we need to define it. However, there is a trick we can use `axis=plt.gca()` to return the current axis being used by `plt`

```python
import matplotlib.ticker as tick
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(12,6))
plt.plot(AAPL.index,AAPL.Close,label="AAPL")
plt.title("AAPL stock price",fontsize=18,pad=10)
plt.legend(fontsize=13)
plt.grid(True)
plt.xlabel("Date",fontsize=15,labelpad=5)
plt.ylabel("Price",fontsize=15,labelpad=5)
#plt.xticks([0,5,10,15,20],["Jan","Feb","Mar","Apr","May"],fontsize=10)
#plt.yticks([-1,0,1],[-10,0,10],fontsize=10)
plt.tight_layout()

def y_fmt(x, y):
    return f"${int(x)}" ## Here we define the format using a fstring
 
plt.gca().yaxis.set_major_formatter(tick.FuncFormatter(y_fmt))

plt.show()
```

### Crash Course on string formatting

By now we have used strings to print messages/tiles etc. There are several ways to embed variable values into a string. The most pythonic way to do that is by using f-strings, which were introduced in Python 3.6.

```python
value = 3.14159
value2=6.325
big = 1234567.891
print("DECIMAL NOTATION:\n") #\n is equivalent to pressing escape and jumping to the next line
print(f"Round to 2 decimal places: {value:.2f}")  # Output: 3.14

print(f"Pad with zeros: {value:06.2f}")  # Output: 003.14

print(f"Align right in 10 spaces: {value:>10.2f}")  # Output: '      3.14'
print(f"Align right in 10 spaces: {value2:>10.3f}")  # Output: '     6.325'

print(f"Align left: {value:<10.2f}")  # Output: '3.14      '
print(f"Align left: {value2:<10.3f}")  # Output: '3.14      

print(f"Use comma as thousands separator: {big:,.2f}")  # Output: '1,234,567.89'
print(f"Use percentage representation: {value/100:.2%}")  
print(f"Use percentage representation: {value:.2f}%")  

print("----------------------------------------")

print("SCIENTIFIC NOTATION:\n") #\n is equivalent to pressing escape and jumping to the next line

num = 0.0001234

# f-string
print(f"{num:.4e}")         # 1.23e-04

print("----------------------------------------")

print("BINARY/HEXADECIMAL NOTATION:\n") #\n is equivalent to pressing escape and jumping to the next line

num = 42

# Binary
print(f"Binary representation of {num} is : {num:b}")           # 101010
```

### likewise, if we are plotting returns, we can use the same technique to present result in % points

```python
import matplotlib.ticker as tick
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(12,6))
plt.plot(AAPL.index,AAPL.Close.pct_change(),label="AAPL")
plt.title("AAPL stock returns",fontsize=18,pad=10)
plt.legend(fontsize=13)
plt.grid(True)
plt.xlabel("Date",fontsize=15,labelpad=5)
plt.ylabel("Returns",fontsize=15,labelpad=5)
#plt.xticks([0,5,10,15,20],["Jan","Feb","Mar","Apr","May"],fontsize=10)
#plt.yticks([-1,0,1],[-10,0,10],fontsize=10)
plt.tight_layout()

def y_fmt(x, y):
    return f"{100*x:.0f}%" ## Here we define the format using a fstring
 
plt.gca().yaxis.set_major_formatter(tick.FuncFormatter(y_fmt))

plt.show()
```

# Default settings for all plots

A nice trick to avoid reperating the same lines of code for each plot is to change the defaults. These are stored under a dictionary in `plt.rcParams` 


```python
plt.rcParams
```

We can play with rcParams to come up with a configurations that we like combining different paramenters, e.g.:

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8, 6]  # width, height in inches
plt.rcParams['axes.grid'] = True       # Show grid by default
plt.rcParams['grid.color'] = 'gray'    # Grid line color
plt.rcParams['grid.linestyle'] = '--'  # Line style of grid (e.g., '-', '--', ':', '-.')
plt.rcParams['grid.linewidth'] = 0.5 
plt.rcParams['legend.loc'] = 'best'         # Default location (e.g., 'upper right', 'lower left')
plt.rcParams['legend.frameon'] = True       # Show legend frame
plt.rcParams['legend.framealpha'] = 0.8     # Frame transparency
plt.rcParams['legend.facecolor'] = 'gray'  # Background color
plt.rcParams['legend.edgecolor'] = 'black'  # Border color
plt.rcParams['legend.fontsize'] = 'Large'  # Font size (e.g., 'small', 'medium', 'large')
# Set default figure background color
plt.rcParams['figure.facecolor'] = 'black'
# Set default axes background color
plt.rcParams['axes.facecolor'] = 'black' 
plt.rcParams['axes.edgecolor'] = 'white'
plt.rcParams['xtick.color'] = 'yellow'
plt.rcParams['ytick.color'] = 'yellow'
```

An interesting modifycation is the default line color cycle. Tipically is a (blue,orange,...) sequence. We can modify this sequenci and it's style using the `cycler` libray 

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,20,100)
plt.plot(x,np.sin(x),label="Sin")
plt.plot(x,np.cos(x),label="Cos")
plt.plot(x,np.sin(x)+np.cos(x),label="Sin+Cos")
plt.legend()
plt.show()
```

```python
from cycler import cycler

# Set custom color cycle globally
plt.rcParams['axes.prop_cycle'] = cycler(color=['red', '#33FFCE', 'pink', '#FF33A8'])+cycler(linestyle=['-', '--', ':','-.'])
```
